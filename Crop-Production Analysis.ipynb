# Productivity trend per year(1997-2015)

```python
table1 = pd.pivot_table(crop,values="Production", index = ['Crop_Year'], aggfunc = 'sum')
table1=table1.sort_values(by = ['Crop_Year'], ascending = False) 
table1
```

```python
plt.style.use('grayscale')
```

```python
table1.Production.plot(figsize = (12,4), kind = 'bar', fontsize = 13, color = 'blue', rot=90)
plt.title('Year-wise Agricultural Productivity',fontsize = 15)
plt.legend(loc = "upper right", fontsize = 18)
plt.xlabel('Year', fontsize = 15)
plt.yticks(rotation=0)
plt.show()
```
